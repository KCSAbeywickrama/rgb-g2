```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
```

```python
with open("..\data\stage2\calibset1.txt") as fl:
    calib1=fl.read().split('\n\n')

with open("..\data\stage2\calibset2.txt") as fl:
    calib2=fl.read().split('\n\n')

with open("..\data\stage2\dataset1.txt") as fl:
    data1=fl.read().split('\n\n')

with open("..\data\stage2\dataset2.txt") as fl:
    data2=fl.read().split('\n\n')

calib1_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in calib1])
calib1_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in calib1])

calib2_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in calib2])
calib2_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in calib2])

data1_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in data1])
data1_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in data1])

data2_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in data2])
data2_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in data2])

print(calib1_ex)
print(calib1_rd)
```

```python
# RED
i=0
plt.figure(figsize=(15, 6))

data1_y=np.split(data1_ex,3,1)[i].reshape(12)
data1_x=np.split(data1_rd,3,1)[i].reshape(12)
plt.scatter(data1_x,data1_y,c="green")

data2_y=np.split(data2_ex,3,1)[i].reshape(36)
data2_x=np.split(data2_rd,3,1)[i].reshape(36)
plt.scatter(data2_x,data2_y,c="blue")


calib1_y=np.split(calib1_ex,3,1)[i].reshape(5)
calib1_x=np.split(calib1_rd,3,1)[i].reshape(5)

calib1_model = np.poly1d(np.polyfit(calib1_x, calib1_y, 1))
calib1_line = np.linspace(290, 700, 100)
plt.plot(calib1_line, calib1_model(calib1_line),c="red")
plt.scatter(calib1_x, calib1_y,c="red")


calib2_y=np.split(calib2_ex,3,1)[i].reshape(5)
calib2_x=np.split(calib2_rd,3,1)[i].reshape(5)

calib2_model = np.poly1d(np.polyfit(calib2_x, calib2_y, 1))
calib2_line = np.linspace(290, 700, 100)
plt.plot(calib2_line, calib2_model(calib2_line),c="black")
plt.scatter(calib2_x, calib2_y,c="black")

plt.show()

print(r2_score(calib1_y, calib1_model(calib1_x)))
print(r2_score(calib2_y, calib2_model(calib2_x)))
```
