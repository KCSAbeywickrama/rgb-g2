```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
```

```python
with open("..\data\stage2\calibset1.txt") as fl:
    calib=fl.read().split('\n\n')

with open("..\data\stage2\dataset1.txt") as fl:
    data1=fl.read().split('\n\n')

with open("..\data\stage2\dataset2.txt") as fl:
    data2=fl.read().split('\n\n')

calib_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in calib])
calib_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in calib])

data1_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in data1])
data1_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in data1])

data2_ex=np.array([tuple(map(int,line.split('\n')[0].split(','))) for line in data2])
data2_rd=np.array([tuple(map(int,line.split('\n')[1].split(','))) for line in data2])

print(calib_ex)
print(calib_rd)
```

```python
# RED
i=0
plt.figure(figsize=(15, 6))

data1_y=np.split(data1_ex,3,1)[i].reshape(12)
data1_x=np.split(data1_rd,3,1)[i].reshape(12)
plt.scatter(data1_x,data1_y,c="green")

data2_y=np.split(data2_ex,3,1)[i].reshape(36)
data2_x=np.split(data2_rd,3,1)[i].reshape(36)
plt.scatter(data2_x,data2_y,c="blue")


calib_y=np.split(calib_ex,3,1)[i].reshape(5)
calib_x=np.split(calib_rd,3,1)[i].reshape(5)

calib_model = np.poly1d(np.polyfit(calib_x, calib_y, 1))
calib_line = np.linspace(290, 700, 100)
plt.plot(calib_line, calib_model(calib_line),c="red")
plt.scatter(calib_x, calib_y,c="red")

plt.show()

print(r2_score(calib_y, calib_model(calib_x)))
```
